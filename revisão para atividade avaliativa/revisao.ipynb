```python
def somar_numeros(a, b):
  # Somar dois números e retorna o resultado
  soma = a + b
  return soma

# Chamando a função com dois números
resultado = somar_numeros(5,3)

# Exibindo o resultado
print(f"A soma é {resultado}")
```

```python
import random
import numpy as np
from matplotlib import pyplot as plt

x = [1,2,3,4]
y = np.random.randint(30,40,4)

plt.ion()
plt.bar(x,y)
plt.pause(4)

plt.cla() # Limpeza dos dados
plt.clf() # Limpeza no geral
y = np.random.randint(10,20,4)
plt.bar(x,y)
plt.pause(4)
```
